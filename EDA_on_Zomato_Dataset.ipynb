## **Exploratory Data Analysis on Zomato Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("himanshupoddar/zomato-bangalore-restaurants")

print("Path to dataset files:", path)
```

```python
# The path to the downloaded dataset files
dataset_path = '/root/.cache/kagglehub/datasets/himanshupoddar/zomato-bangalore-restaurants/versions/1'

# Construct the full path to the CSV file
csv_file_path = f'{dataset_path}/zomato.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame to verify
display(df.head())
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

**Dropping Duplicates**

```python
df.drop_duplicates(inplace = True)
df.shape
```

**Cleaning Rate Column**

```python
df['rate'].unique()
```

**Removing "NEW" , "-" and "/5" from Rate Column**


```python
def handlerate(value):
  if value == 'NEW' or value == '-' :
    return np.nan
  else:
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

**Filling Null Values in Rate Column with Mean**

```python
df['rate'] = df['rate'].fillna(df['rate'].mean())
df['rate'].isnull().sum()
```

```python
df.info()
```

**Dropping Null Values**

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)' : 'Costfor_2' , 'listed_in(type)' : 'Type'} , inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df = df.drop(['listed_in(city)'],axis = 1)
df.head()
```

```python
df['Costfor2'].unique()
```

```python
def handlecomma(value):
  value = str(value)
  if ',' in value :
    value = value.replace(',','')
    return float(value)
  else:
    return float(value)

df['Costfor2'] = df['Costfor2'].apply(handlecomma)
df['Costfor2'].unique()
```

```python
df.head()
```

```python
df['rest_type'].unique()
```

**Cleaning Rest_type column**

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

**Making Rest Types less than 1000 in frequency as others**

```python
def handle_rest_type(value):
  if value in rest_types_lessthan1000:
    return 'others'
  else:
    return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['rest_type'].unique()
```

```python
df.head()
```

**Cleaning Location Column**

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]
```

```python
def handle_location(value):
  if value in location_lessthan300:
    return 'others'
  else:
    return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

**Cleaning cuisines Column**

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def handle_cuisines(value):
  if value in cuisines_lessthan100:
    return 'others'
  else:
    return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
